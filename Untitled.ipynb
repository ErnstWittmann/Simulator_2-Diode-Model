```python
from Panel_Simulation import Panel_Simulator
from UsefullMiniFunctions import plt_U_I
import matplotlib.pyplot as plt
from tqdm import tqdm
import multiprocessing as mp
import pandas as pd
import pandas as pd

def module_discription(waver_per_string=10,
                      string_per_module = 3,
                       I_SC = 5.17, #A
                       shunt_resistance = 50, #Ohm,
                       reverse_leakage_current_1 = 1e-9, #nA,
                       N1 = 2,
                       reverse_leakage_current_2 = 1e-9, #nA,
                       N2 = 1,
                       serial_resistance = 0.001 #Ohm
                      ):
    waver_list = np.ones([string_per_module, waver_per_string, 7])
    waver_list[:, :, 0] = I_SC  # generator_current
    waver_list[:, :, 1] = shunt_resistance
    waver_list[:, :, 2] = reverse_leakage_current_1
    waver_list[:, :, 3] = N1
    waver_list[:, :, 4] = reverse_leakage_current_2
    waver_list[:, :, 5] = N2
    waver_list[:, :, 6] = serial_resistance
    return waver_list
```

```python
#
print("start")
from bayes_opt import BayesianOptimization

waver_per_string=20
string_per_module = 3
N1 = 2
N2 = 1
I_SC = 9.35 #A

#-----------------------------------------------

reverse_leakage_current1 = 3.6e-11#nA,

serial_resistance = 0.005#462 #Ohm
shunt_resistance = 150 #Ohm,

bypass_IS=1e-5
bypass_N=2
bypass_RS=0.10
bypass_BV=40 



# Bounded region of parameter space
pbounds = {'I_SC': (5, 11), 'I_SC1': (0.6, 1), 'I_SC2': (0.6, 1), 'I_SC3': (0.6, 1),
           'bypass_N':(1, 6), "bypass_BV":(10,100),"bypass_RS":(0.0001,0.1),'bypass_IS':(1e-14,1e-9),
          'serial_resistance':(0.00000001,1),'sr1':(0.00000001,1),'sr2':(0.00000001,1),'sr3':(0.00000001,1),
           "shunt_resistance":(0,1000),"shr1":(0,1000),"shr2":(0,1000),"shr3":(0,1000),
           "reverse_leakage_current":(1e-11,5e-11)
          }



def black_box_function(I_SC, I_SC1,I_SC2,I_SC3,
                       bypass_N,bypass_BV,bypass_RS,bypass_IS,
                       serial_resistance,sr1,sr2,sr3,
                       shunt_resistance,shr1,shr2,shr3,
                       reverse_leakage_current):
    module = module_discription(  
                                               waver_per_string,
                                               string_per_module,
                                               I_SC,
                                               shunt_resistance,
                                               reverse_leakage_current,
                                               N1,
                                               reverse_leakage_current,
                                               N2,
                                               serial_resistance
                                            )
    module[0,0,0]=I_SC*I_SC1
    module[1,0,0]=I_SC*I_SC2
    module[2,0,0]=I_SC*I_SC3
    module[0,0,6]=sr1
    module[1,0,6]=sr2
    module[2,0,6]=sr3
    module[0,0,1]=shr1
    module[1,0,1]=shr2
    module[2,0,1]=shr3
    
    
    
    df = pd.read_csv(r"G:\Praktikum_07-01-2024\823001_1000W.csv")
    sim = Panel_Simulator()

    module_circuit = sim.create_module(module,bypass_IS=bypass_IS, bypass_N=bypass_N, bypass_RS=bypass_RS, bypass_BV=bypass_BV)

    u,i = sim.analyse_module( U_start = 0, U_end = 42, resolution=4000, temperature=25, nominal_temperature=25)

    simulation = pd.DataFrame()
    simulation["Voltage(V)"] = u
    simulation["sim_I(A)"] = i

    df = df.sort_values("Voltage(V)")
    df = pd.merge_asof(df,simulation, on="Voltage(V)")
    df = df.dropna()
    
    return np.sum(np.abs(df["Current(A)"]-df["sim_I(A)"]))




optimizer = BayesianOptimization(
    f=black_box_function,
    pbounds=pbounds,
    random_state=1,
    verbose=0,
)
optimizer.maximize(
    init_points=10000,
    n_iter=50,
)
pred_z = np.array([z["target"] for z in optimizer.res])
index = np.argmin(pred_z)
rp = optimizer.res[index]["params"]
print(optimizer.res[index])



module = module_discription(  
                           waver_per_string,
                           string_per_module,
                           rp["I_SC"],
                           shunt_resistance,
                           rp["reverse_leakage_current"],
                           N1,
                           rp["reverse_leakage_current"],
                           N2,
                           rp["serial_resistance"]
                        )
sim = Panel_Simulator()
module_circuit = sim.create_module(module,bypass_IS=rp["bypass_IS"], 
                                          bypass_N=rp["bypass_N"], 
                                          bypass_RS=rp["bypass_RS"], 
                                          bypass_BV=rp["bypass_BV"])

u,i = sim.analyse_module( U_start = 0, U_end = 42, resolution=4000, temperature=25, nominal_temperature=25)
plt.plot(u,i, label="IV-Curve")
plt.plot(df["Voltage(V)"], df["Current(A)"],color="red")
plt.ylim(0,10)
plt.xlim(0,42)
plt.legend()
plt_U_I()
plt.show()
```
